# Plot of Total P against Drainage Area

```r
total.P.plot<-plot(ResultMeasureValue~DrainageAreaMeasure_MeasureValue, 
                   data=subset(landuse.p, ResultSampleFractionText=="Total"), 
                   type="p", xlab="Drainage Area (sq. miles)", ylab="Total Phosphorus mg/L")
total.P.plot
```

# Plot of TDP against Drainage Area

```r
TDP.plot<-plot(ResultMeasureValue~DrainageAreaMeasure_MeasureValue, 
               data=subset(landuse.p, ResultSampleFractionText=="Dissolved"), 
               type="p", xlab="Drainage Area (sq. miles)", ylab="Total Dissolved Phosphorus mg/L")
TDP.plot
```

# Plot of TDP against Drainage Area by HUC

```r
TDP.HUC.plot<-xyplot(ResultMeasureValue~DrainageAreaMeasure_MeasureValue|HUC8, 
                     data=subset(landuse.p, ResultSampleFractionText=="Dissolved"), 
                     type="p", xlab="Drainage Area (sq. miles)", ylab="Total Dissolved Phosphorus mg/L")
TDP.HUC.plot
```

# Timeplot of TDP by HUC

```r
TDP.HUC.timeplot<-xyplot(ResultMeasureValue~date|HUC8, 
                     data=subset(landuse.p, ResultSampleFractionText=="Dissolved"), 
                     type="p", xlab="Date", ylab="Total Dissolved Phosphorus mg/L")
TDP.HUC.timeplot
```

# Timeplot of Total P by HUC

```r
TotalP.HUC.timeplot<-xyplot(ResultMeasureValue~date|HUC8, 
                     data=subset(landuse.p, ResultSampleFractionText=="Total"), 
                     type="p", xlab="Date", ylab="Total Phosphorus mg/L")
TotalP.HUC.timeplot
```
